```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from pandas.core.common import random_state
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('score.csv')
df.head()
```

```python
df.info()
```

```python
plt.title('Grades After Hours Of Study')
sns.distplot(df['Hours'])
plt.show()
```

```python
plt.scatter(df['Hours'], df['Scores'], color = 'lightcoral')
plt.title('Hours Of Study vs Grades')
plt.xlabel('Hours Of Study')
plt.ylabel('Grades')
plt.box(False)
plt.show()
```

```python
X = df[['Hours']]  
y = df['Scores'] 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_pred_test = regressor.predict(X_test)
y_pred_train = regressor.predict(X_train) 
```

```python
plt.scatter(X_train, y_train, color = 'lightcoral')
plt.plot(X_train, y_pred_train, color = 'firebrick')
plt.title('Hours Of Study vs Grades (Training Set)')
plt.xlabel('Hours Of Study')
plt.ylabel('Grades')
plt.legend(['X_train/Pred(y_test)', 'X_train/y_train'], title = 'Hours Of Study/Grades', loc='best', facecolor='white')
plt.box(False)
plt.show()
```

```python
plt.scatter(X_test, y_test, color = 'lightcoral')
plt.plot(X_train, y_pred_train, color = 'firebrick')
plt.title('Hours Of Study Vs Grades (Test Set)')
plt.xlabel('Hours Of Study')
plt.ylabel('Grades')
plt.legend(['X_train/Pred(y_test)', 'X_train/y_train'], title = 'Hours Of Study/Grades', loc='best', facecolor='white')
plt.box(False)
plt.show()
```
